```python
def calc_perm(i):
    if (i.size == 0):
        return 0
    return np.average(i)
```

```python
guest_1_labels = ["Q39_"+str(i) for i in range(1,11)] # Definitely have security and privacy concerns when sharing
guest_2_labels = ["Q55_"+str(i) for i in range(1,11)] # Some security and privacy concerns when sharing
guest_3_labels = ["Q56_"+str(i) for i in range(1,11)] # Definitely do not have security and privacy concerns when sharing

guest_all_labels = guest_1_labels + guest_2_labels + guest_3_labels # all 
# print(guest_all_labels)
```

```python
guest_1_perm = []
df_1 = df.loc[:, df.columns.intersection(guest_1_labels)]
for i in guest_1_labels:
    norm_guest_group = np.array(df_1[i].dropna() / 5)
    guest_1_perm.append(calc_perm(norm_guest_group))

k2, p = stats.normaltest(guest_1_perm)
print(p)

guest_2_perm = []
df_2 = df.loc[:, df.columns.intersection(guest_2_labels)]

for i in guest_2_labels:
    norm_guest_group = np.array(df_2[i].dropna() / 5)
    guest_2_perm.append(calc_perm(norm_guest_group))
    
k2, p = stats.normaltest(guest_2_perm)
print(p)

guest_3_perm = []
df_3 = df.loc[:, df.columns.intersection(guest_3_labels)]

for i in guest_3_labels:
    norm_guest_group = np.array(df_3[i].dropna() / 5)
    guest_3_perm.append(calc_perm(norm_guest_group))
    
k2, p = stats.normaltest(guest_3_perm)
print(p)

result = stats.kruskal(guest_1_perm, guest_2_perm)
print('kruskal-wallis one way, f and p values for data science notebooks', result[1] < 0.016)

result = stats.kruskal(guest_2_perm, guest_3_perm)
print('kruskal-wallis one way, f and p values for data science notebooks', result[1] < 0.016)

result = stats.kruskal(guest_3_perm, guest_1_perm)
print('kruskal-wallis one way, f and p values for data science notebooks', result[1] < 0.016)
```
